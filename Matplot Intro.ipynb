***For this code ,we will take the dataset of COVID-19 cases***

**1.Simple data plotting using plot() function**

```python
# Load the required libraries 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# Load the COVID 19 dataset
df = pd.read_csv(r"C:\Users\Aniket Chauhan\Desktop\datasets_557629_1349299_covid_19_india.csv")
df.head(2)
```

```python
# Take the x axis for the plot as Death 
# Take the y axis for the plot as Cured 

x = df[['Deaths']].to_numpy()
y = df[['Cured']].to_numpy()
```

```python
# Create plot using the plot() function
plt.xlabel('Deaths')
plt.ylabel('Cured')
plt.title('Simple data plotting using plot() function')
plt.plot(x,y) 
plt.show() 
```

```python
# Data Grouping Based on States and then plotting
state = df.groupby('State/UnionTerritory')  
state = state.agg(np.sum)

# Take the x axis for the plot as Date 
# Take the y axis for the plot as Cured 

x = state [['Deaths']].to_numpy()
y = state [['Cured']].to_numpy()

# Create plot using the plot() function
plt.xlabel('Deaths')
plt.ylabel('Cured')
plt.title('Data Grouping Based on States')
plt.plot(x,y) 
plt.show()
```

**2. Scatter data plotting using scatter() function**

#These plot are often useful when the number of datapoints are less in number

```python
# Create plot using the scatter() function
plt.xlabel('Deaths')
plt.ylabel('Cured')
plt.title('Simple Scatter plot')
plt.scatter(x, y) 
plt.show() 
```

```python
date = df.groupby('Date')  
date = date.agg(np.sum)

# Take the x axis for the plot as Date 
# Take the y axis for the plot as Cured 

x = date [['Deaths']].to_numpy()
y = date [['Cured']].to_numpy()

# Create plot using the plot() function
plt.xlabel('Deaths')
plt.ylabel('Cured')
plt.title('Scatter Plot based Data Grouping Based on Dates')
plt.scatter(x, y) 
plt.show() 
```
