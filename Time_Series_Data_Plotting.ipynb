# Time Series Data Plotting

1. The time series data is a data that varies with the data and time.
2. Time series data is used in various projects such as stock market prediction , variation of pollutants in the city over time,etc.
3. The most common way to plot time series data is to use line plot where the data is indexed on y axis and time is indexed on X axis.
4. Time series data is a data whose valeus are the function of time i.e data collects depends on time.


**For this code ,we will take the dataset of Air Quality Data in India (2015 - 2020)**

About data set :
1. The dataset contains concentration of various air pollutants at hourly and daily level of various stations across multiple cities in India.
2. The concentration is measured in ppm(parts per million)

```python
# Load the required libraries 
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
```

```python
# Load the dataset
dd = pd.read_csv(r"C:\Users\Aniket Chauhan\Desktop\AI Vidya\city_day.csv")
dd.head(2)
```

```python
# We will create the list of pollutants which are needed to be plot
dd_col = ['NO', 'NO2', 'NOx', 'NH3', 'CO','SO2', 'O3']
print(dd_col)
```

```python
# Taking the data for the city Ahmedabad
Ahemdabad= dd.loc[dd['City'] == 'Ahmedabad']
Ahemdabad.head(10)

# NaN means not a number
```

```python
# Convert the Date Column of the Ahemdabad dataframe to Standard DateTime format
Ahemdabad['Date']= pd.to_datetime(Ahemdabad['Date']) 

# Print the shape of Ahemdabad 
print(Ahemdabad.shape)

# Create list of pollutants names whihc are removed form the Ahemdabad data frame
c =  ['City' , 'PM2.5' , 'PM10','Benzene' , 'Toluene', 'Xylene'  ,'AQI' ,'AQI_Bucket']
```

```python
# The the columnn and the sort the values on the basis of Date column
Ahemdabad_new  = Ahemdabad.drop(c, axis=1)
Ahemdabad_new= Ahemdabad_new.sort_values('Date')
```

```python
Ahemdabad_new.head(5)
```

```python
# Reset the index(row number order )
print(Ahemdabad_new.index)
Ahemdabad_new= Ahemdabad_new.reset_index()

# Set the index as Date
Ahemdabad_new = Ahemdabad_new.set_index('Date')
print(Ahemdabad_new.index)
```

```python
NO = Ahemdabad_new['NO']
NO.plot(figsize=(15, 6))
plt.show()
```

```python
# Now we resample the data on weekly basis and plot it
# Considering the pollutant NO 
NO_W = Ahemdabad_new['NO'].resample('2W').mean()
NO_W.plot(figsize=(15, 6))
plt.show()
```

```python
# To remove these null values , we will fill these null values the mean of the overall column value as 
for i in Ahemdabad_new.columns:
    a = Ahemdabad_new[i].mean()
    Ahemdabad_new[i].replace(np.nan , a,inplace =  True)
```

```python
NO = Ahemdabad_new['NO']
NO.plot(figsize=(15, 6))
plt.show()
```

```python
# Now we resample the data on weekly basis after removal of null values and plot it
# Considering the pollutant NO 
NO_W = Ahemdabad_new['NO'].resample('2W').mean()
NO_W.plot(figsize=(15, 6))
plt.show()
```
